```python
import pandas as pd
import plotly as plt
import os
import glob
import plotly.express as px
import numpy as np
# Goal : Histogramme des personnes ayant été reçus ou non et dans quelle formation (BTS / Droit / Art / Inge / Commerce / Pro / Autre ?)
# 
pd.set_option('display.max_columns', None)


def split_apb(df):
    grouped = df.groupby(df.Session)
    print(grouped)
    return grouped.get_group(2016),grouped.get_group(2017)
    
def open_csv(path, separator):
    return pd.read_csv(path, sep=separator)


def sum_candidates_accepted(df, column_name, apb):
    column = df[[column_name]]
    if apb:
        column = column[column[column_name].str.contains('inconnu')==False]
        column = column.astype('int64')
    return column.sum()[column_name]

#def remove_NaNs_and_unknown(dataframe):
#    for x in dataframe:
#        x = x.replace('inconnu', '0')
#        if type(x) == 'int64':
#            x = x.replace(np.nan, 0)
#        if type(x) == 'float64':
#            x = x.astype('int64')
#            x = x.replace(np.nan, 0)
#    return dataframe

path = os.path.join(os.getcwd(), 'data')
path = os.path.join(path, '*.csv')

files = glob.glob(path)

dataframe = pd.concat((pd.read_csv(i, sep = ';') for i in files))
#dataframe.reset_index(drop = True, inplace = True)

dataframe = remove_NaNs_and_unknown(dataframe)

dataframe.rename(columns = {'Effectif total des candidats ayant accepté la proposition de l’établissement (admis)' : 'Candidats admis parcoursup'}, inplace = True)
dataframe.rename(columns = {'Effectif total des candidats admis' : 'Candidats admis APB'}, inplace = True)

dataframe['Candidats admis APB'] = dataframe['Candidats admis APB'].replace('inconnu', '0')
dataframe['Candidats admis parcoursup'] = dataframe['Candidats admis parcoursup'].replace('inconnu', '0')

dataframe['Candidats admis APB'] = dataframe['Candidats admis APB'].replace(np.nan, 0)
dataframe['Départements'] = dataframe['Départements'].replace(np.nan, 0)
dataframe['Candidats admis parcoursup'] = dataframe['Candidats admis parcoursup'].replace(np.nan, '')
dataframe['Filières de formations'] = dataframe['Filières de formations'].replace(np.nan, '')
dataframe['Filière de formation'] = dataframe['Filière de formation'].replace(np.nan, '')
dataframe['Filière de formation très détaillée'] = dataframe['Filière de formation très détaillée'].replace(np.nan, '')


dataframe['Admis'] = dataframe['Candidats admis APB'] + dataframe['Candidats admis parcoursup']
dataframe['Filière de formation'] = dataframe['Filière de formation'] + dataframe['Filières de formations']


dataframe['Admis'] = dataframe['Admis'].astype('int64')
#dataframe2 = dataframe[["Session", "candidats admis parcoursup", "candidats admis APB"]]
dataframe
```

```python
df_parcousup_2019 = open_csv('data/parcoursup_2019.csv', ';')
somme_2019 = sum_candidates_accepted(df_parcousup_2019, 'Effectif total des candidats ayant accepté la proposition de l’établissement (admis)', False)
somme_2019
```

```python
df_parcousup_2021 = open_csv('data/parcoursup_2021.csv', ';')
somme_2021 = sum_candidates_accepted(df_parcousup_2021,'Effectif total des candidats ayant accepté la proposition de l’établissement (admis)', False)
somme_2021
```

```python
df_apb = open_csv('data/apb.csv', ';')

df_apb_2016, df_apb_2017 = split_apb(df_apb)
somme_apb_2016 = sum_candidates_accepted(df_apb_2017,'Effectif total des candidats', True)
somme_apb_2017 = sum_candidates_accepted(df_apb_2016,'Effectif total des candidats admis', True)
somme_apb_2016
```
